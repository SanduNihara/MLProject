```python
df
```

```python
df_nulldropped = df.dropna(how = "all")
df_nulldropped
```

```python
df_fillwithzero = df.fillna(0)
df_fillwithzero
```

```python
df_forwardfilled = df.fillna(method = 'ffill')
df_forwardfilled
```

```python
df_backwardfilled = df.fillna(method = 'bfill')
df_backwardfilled
```
